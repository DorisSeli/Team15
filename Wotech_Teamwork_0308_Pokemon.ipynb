```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pokemon_df = pd.read_csv('/content/Pokemon.csv')
pokemon_df.head(n=10)
```

```python
#How many Pokemons have Type 1 == Water?

Type1_Water = pokemon_df[pokemon_df['Type 1'] == 'Water']
total_pokemon = len(pokemon_df)
count_Type1_Water = len(Type1_Water)
percentage_Type1_Water = (count_Type1_Water / total_pokemon) * 100
print(count_Type1_Water)
print(percentage_Type1_Water)
```

```python
# What is the maximum 'Speed' value? What is the minimum 'Speed' value?
#What is the difference between max and min 'Speed'?

max_Speed = pokemon_df['Speed'].max()
min_Speed = pokemon_df['Speed'].min()
print(max_Speed)
print(min_Speed)

speed_difference = max_Speed - min_Speed
print(speed_difference)

print('The maximum "Speed" value is:', {max_Speed}, 'and the minimum "Speed" is:', {min_Speed}, '. The difference between max and min is:', {speed_difference}, '.')
```

```python
# Filter the DataFrame to include only the Pokémon with 'Speed' >= 80. How many Pokémon meet this criterion? Display this DataFrame using your preferred visualization method.

speed_over80_pokemon = pokemon_df[pokemon_df['Speed'] >= 80]
print(len(speed_over80_pokemon))
```

```python
# Total speed histogram
value = plt.hist(pokemon_df['Speed'], bins=range(0, 200, 10), color='teal', edgecolor='black', alpha=0.7)

plt.xlabel('Speed')
plt.ylabel('Number of pokemons')

plt.axvline(pokemon_df['Speed'].mean(), linestyle='dashed', color='red')
plt.grid(linestyle='--', alpha=0.7)
```

```python
#Speed histogram for Speed over 80
values, bins, patches = plt.hist(speed_over80_pokemon['Speed'], bins=range(80, 200, 10), color='teal', edgecolor='black', alpha=0.7)

plt.xlabel('Speed')
plt.ylabel('Number of pokemons')

for value, bin in zip(values, bins[:-1]):
    plt.text(bin + (bins[1] - bins[0]) / 2, value, int(value), ha='center', va='bottom')

plt.axvline(speed_over80_pokemon['Speed'].mean(), linestyle='dashed', color='red')
plt.grid(linestyle='--', alpha=0.7)
```

```python
#(DIFFICULT) Find Pokémon with the longest name (excluding spaces)? What is this pokemons name?
longest_name = pokemon_df['Name'].apply(lambda x: x.replace(" ", "")).str.len().idxmax()
longest_name_is = pokemon_df['Name'][longest_name]
longest_name_len = len(longest_name_is.replace(" ", ""))

print(f"The longest name in Pokemon dataframe is: '{longest_name_is}' with length {longest_name_len}.")

```
