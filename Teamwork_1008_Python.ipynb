```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pokemon_df = pd.read_csv('/content/Pokemon.csv')
pokemon_df.head(n=5)
```

```python
#1. Create two DataFrame Grass and Water
grass_df = pokemon_df[pokemon_df['Type 1'] == 'Grass']
grass_df.head(n=5)
water_df = pokemon_df[pokemon_df['Type 1'] == 'Water']
water_df.head(n=5)
```

```python
# Create the regression plots for each (Grass and Water)
#Regression plots in the same graph
# Calculate the Pearson correlation for each DataFrame (variables: Attack and Defense)


sns.regplot(
    data=grass_df, x='Attack', y="Defense",
    ci=99, marker="x", color=".3", line_kws=dict(color="r"),
)
correlation_grass = grass_df['Attack'].corr(pokemon_df['Defense'])
print('Pearson correlation coefficient:', correlation_grass)

sns.regplot(
    data=water_df, x='Attack', y="Defense",
    ci=99, marker="x", color=".3", line_kws=dict(color="r"),
)
correlation_water = water_df['Attack'].corr(pokemon_df['Defense'])
print('Pearson correlation coefficient:', correlation_water)
```
